# Customer Churn Prediction
Notebook demonstrating EDA, model training, and evaluation

```
import pandas as pd
churn = pd.read_csv('churn.csv')
churn.head()
```

```
from sklearn.model_selection import train_test_split
X = churn[['tenure_months','monthly_charges','total_charges','support_tickets']]
y = churn['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
len(X_train)
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
preds = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, preds))
print('Confusion matrix:', confusion_matrix(y_test, preds))
```

```
import joblib
joblib.dump(model, 'rf_churn_model.joblib')
print('Model saved')
```
